Eulerian and Lagrangian observations
-----
The goal of these notes is to better understand the difference between Eulerian and Lagrangian observations of currents in the upper ocean. 

We'll begin with some simple computations to show how these different reference frames modulate our measurements. 

Things to keep in mind
-----
* The main difference between the two frames it the Doppler shift induced by the Stokes drift. If both frames had some mean current acting on them that was the same, this would not be an issue. The issue is that the Eulerian mean is zero for water waves, while the Lagrangian mean is the Stokes drift. 

* Nonlinearities can contribute significantly to the Stokes drift (remember the wave tank data). 

* The Lagrangian frame removes the second order nonlinear harmonic

* The Lagrangian frame allows for vorticity. 


Wave spectra
-----
We begin by considering the difference of wave spectra due to Eulerian (fixed) and Lagrangian (moving with the flow) measurements. Here, we are assuming we are recording temporal angular frequency of the waves, $\omega$. We start with 1 dimensional considerations. 

The simplest way to see what is going on is to consider the following (which is from Longuet-Higgins 1986). Consider the period of a uniform train of waves of wavelength $L$, moving at speed $c$. The Eulerian period at a fixed place is 
$$T_E = \frac{L}{c}.$$

If we were moving with the flow, there is also a mean velocity that affects the speed at which the wave seems to move, and that is $U$, here taken to be the Stokes drift. Then, the Lagrangian period is 
$$T_L = \frac{L}{c-U}.$$

The difference between $T_L$ and $T_E$ is the Doppler shift. For steep waves, the measure of $(T_L-T_E)/T_E$ can be significant, almost $0.4$. 

Note, if both systems were being advected by some other velocity $V$, then the denominator of $T_E$ is $c-V$ and that of $T_L$ is $c-U-V$. 


The dispersion relationship for these waves in a lab (i.e. Eulerian) frame is 
$$\omega = U k +\sqrt{gk}$$. 

Spectra
----
In the Lagrangian frame, we are measuring amplitudes $a(\sigma)$ where $\sigma = \sqrt{gk}$. The spectral density is proportional to 
$$|a|^2 \delta \sigma.$$
Meanwhile, in the Eulerian frame we measure amplitudes $b(\omega)$, so that the spectral density in this frame is proportional to
$$|b|^2 \delta \omega.$$
  

Consider a monochromatic wave, with Lagrangian angular frequency $\sigma_0$. The spectra looks like a delta-function at this frequency. Meanwhile, in the Lagrangian frame this delta-function now occurs at some other frequency $\sigma_0+Uk_0$, i.e. it is shifted to the right for $U>0$. We can look at this in more detail now. 

```matlab
%% plot comparing Eulerian and Lagrangian spectra. 
% assume the general form is something simple. 
sigma_L = 0 : 1e-2 : 2.5; 
S_L = sigma_L.*exp(-sigma_L.^4);
U = 1; % Set current
g = 9.81; 
k = sigma_L.^2/g;
sigma_E = U*k+sigma_L; 
imatlab_export_fig('print-png')
clf
plot(sigma_L,S_L)
hold on
plot(sigma_E,S_L,'--')
xlabel('\sigma')
ylabel('S(\sigma)')
    set(gcf, 'Position',  [100, 100, 300, 200])
    set(gcf,'color','w');
```
